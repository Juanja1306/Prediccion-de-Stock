# 4. Feature Engineering temporal y de negocio

````python
# 4.1 – Variables temporales a partir de 'dt'
df['día_de_la_semana'] = df['dt'].dt.weekday             # 0 = lunes, …, 6 = domingo
df['mes'] = df['dt'].dt.month                            # 1–12
df['es_fin_de_semana'] = df['día_de_la_semana'].isin([5, 6]).astype(int)

# 4.2 – Ratio de negocio: sale_amount / stock_hour6_22_cnt
#      Evitamos división por cero reemplazando 0 → NaN temporalmente
df['ratio_sale_stock'] = df['sale_amount'] / df['stock_hour6_22_cnt'].replace({0: np.nan})
df['ratio_sale_stock'] = df['ratio_sale_stock'].replace([np.inf, -np.inf, np.nan], 0.0)
````

````python
# 4.3 – Variables categóricas: one-hot para categorías de producto
# # – Identificar las 5 categorías más frecuentes en first_category_id
# 4.3.1 – One-hot en first_category_id (usar strings para evitar error de Categorical)
top_first_cats = df['first_category_id'].value_counts().nlargest(5).index.tolist()
# Convertir a string para no tener problema al asignar "OTRO"
df['first_category_id_str'] = df['first_category_id'].astype(str)
df['first_cat_top'] = df['first_category_id_str'].where(
    df['first_category_id_str'].isin([str(x) for x in top_first_cats]),
    other='OTRO'
)
dummies_first = pd.get_dummies(df['first_cat_top'], prefix='cat1')
df = pd.concat([df, dummies_first], axis=1)
df.drop(columns=['first_category_id_str'], inplace=True)  # Eliminar columna auxiliar
````

````python
# 4.3.2 – One-hot en second_category_id (mismo procedimiento)
top_second_cats = df['second_category_id'].value_counts().nlargest(5).index.tolist()
df['second_category_id_str'] = df['second_category_id'].astype(str)
df['second_cat_top'] = df['second_category_id_str'].where(
    df['second_category_id_str'].isin([str(x) for x in top_second_cats]),
    other='OTRO'
)
dummies_second = pd.get_dummies(df['second_cat_top'], prefix='cat2')
df = pd.concat([df, dummies_second], axis=1)
df.drop(columns=['second_category_id_str'], inplace=True)  # Eliminar columna auxiliar
````

````python
# 4.4 – Justificación breve y listado de nuevas variables
nuevas_feats = [
    'día_de_la_semana', 'mes', 'es_fin_de_semana',
    'ratio_sale_stock'
] + list(dummies_first.columns) + list(dummies_second.columns)
print("\nNuevas columnas generadas en Persona B:")
for col in nuevas_feats:
    print(f"• {col}")
````

* día_de_la_semana: poten­cia captura de estacionalidad semanal (picos fines de semana).
* mes: modela estacionalidad mensual (temporadas altas/bajas en ciertos meses).
* es_fin_de_semana: indicador binario (1 si es sábado/domingo), detecta variación fin de semana.
* ratio_sale_stock: mide rotación de inventario; alto ratio puede anticipar faltantes o reaprovisionamiento.
* cat1_<…>: one-hot para top 5 first_category_id + “OTRO”. Permite al modelo aprender dependencias según categoría de primer nivel.
* cat2_<…>: one-hot para top 5 second_category_id + “OTRO”. Captura subcategorías relevantes.

# 5. Agrupación y generación de secuencias por producto

````python
# 5.1 – Ordenar DataFrame por product_id y dt
df = df.sort_values(by=['product_id', 'dt']).reset_index(drop=True)

# 5.2 – Definir tamaño de ventana (n = 14 días)
n = 14

# 5.3 – Construir lista donde se acumulen las filas de cada ventana
ventanas = []
````

````python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=pd.errors.PerformanceWarning)
````

````python

n = 14

# 1) Asegurarse de que 'dt' sea datetime y ordenar por producto y fecha
df['dt'] = pd.to_datetime(df['dt'])
df = df.sort_values(['product_id', 'dt']).reset_index(drop=True)

# 2) Crear las 14 columnas de lag para 'sale_amount'
#    De esta forma, '* sale_t-1' = venta de 1 día atrás, ..., '* sale_t-14' = venta de 14 días atrás
for lag in range(1, n+1):
    df[f'* sale_t-{lag}'] = df.groupby('product_id')['sale_amount'].shift(lag)

# 3) Crear las 14 columnas de lag para cada feature temporal/numérica
features_temporales = ['día_de_la_semana', 'mes', 'es_fin_de_semana', 'ratio_sale_stock']
for feat in features_temporales:
    for lag in range(1, n+1):
        df[f'{feat}_t-{lag}'] = df.groupby('product_id')[feat].shift(lag)

# 4) Crear las 14 columnas de lag para cada columna one-hot de categoría
#    (asegúrate de que ya existan en 'df' las columnas de dummies, p.ej. dummies_first.columns y dummies_second.columns)
cats_onehot = list(dummies_first.columns) + list(dummies_second.columns)
for cat_col in cats_onehot:
    for lag in range(1, n+1):
        df[f'{cat_col}_t-{lag}'] = df.groupby('product_id')[cat_col].shift(lag)

# 5) Crear el target: 'venta del día siguiente' (shift -1)
df['target_sale_next'] = df.groupby('product_id')['sale_amount'].shift(-1)

# 6) Listar todas las columnas que deben estar completas para poder entrenar
lag_columns = []
# - Lags de venta
lag_columns += [f'* sale_t-{lag}' for lag in range(1, n+1)]
# - Lags de features temporales
for feat in features_temporales:
    lag_columns += [f'{feat}_t-{lag}' for lag in range(1, n+1)]
# - Lags de one-hot categorías
for cat_col in cats_onehot:
    lag_columns += [f'{cat_col}_t-{lag}' for lag in range(1, n+1)]
# - Target
lag_columns += ['target_sale_next']

# 7) Eliminar todas las filas que tengan NaN en cualquiera de esas columnas
df_secuencias = df.dropna(subset=lag_columns).copy()

# 8) Opcionalmente, definir la columna 'fecha_final_ventana' igual a 'dt' (último día de la ventana)
df_secuencias['fecha_final_ventana'] = df_secuencias['dt']

# 9) Seleccionar únicamente las columnas finales que servirán al modelo:
#    ['product_id', 'fecha_final_ventana', 'target_sale_next'] + todos los lag_columns
columnas_finales = ['product_id', 'fecha_final_ventana', 'target_sale_next'] + lag_columns
df_secuencias = df_secuencias[columnas_finales]
````

A continuación se presenta un listado detallado de **todos los lag features** generados y de la variable **`target_sale_next`**, indicando para cada uno:

1. **Fórmula de cálculo** (cómo se generó en el notebook)
2. **Propósito** (por qué aporta información al modelo)

---

### 1. Lag Features de `sale_amount`

Para cada entero $\ell = 1,2,\dots,14$:

* **Nombre de la columna**:

  ```
  * sale_t-ℓ
  ```

  donde “ℓ” indica el número de días atrás (por ejemplo: `* sale_t-1`, `* sale_t-2`, …, `* sale_t-14`).

* **Fórmula**:

  ```python
  df[f"* sale_t-{ℓ}"] = df.groupby("product_id")["sale_amount"].shift(ℓ)
  ```

  Es decir, para cada registro, toma el valor de `sale_amount` del mismo `product_id` que ocurrió ℓ días antes.

* **Propósito**:

  * Capturar la tendencia reciente de ventas de cada producto: si un artículo vendió mucho hace 1, 2 o hasta 14 días, esto suele influir en la predicción de lo que venderá en el futuro.
  * Ayuda a modelar:

    * **Estacionalidad de corto plazo** (por ejemplo, si ℓ = 7, refleja la venta de hace exactamente una semana).
    * **Efecto autocorrelacional** (venta de hoy depende de venta de ayer, anteayer, etc.).
  * En un modelo de series temporales o de Machine Learning, estos lags sirven como variables independientes que contienen la “historia” de ventas recientes.

---

### 2. Lag Features de `día_de_la_semana`

Para cada entero $\ell = 1,2,\dots,14$:

* **Nombre de la columna**:

  ```
  día_de_la_semana_t-ℓ
  ```

  (por ejemplo: `día_de_la_semana_t-1`, `día_de_la_semana_t-2`, …, `día_de_la_semana_t-14`).

* **Fórmula**:

  ```python
  df[f"día_de_la_semana_t-{ℓ}"] = df.groupby("product_id")["día_de_la_semana"].shift(ℓ)
  ```

  Donde `día_de_la_semana` ya había sido calculado como `df['dt'].dt.weekday` (0 = lunes, …, 6 = domingo) para cada registro original.

* **Propósito**:

  * Indicar cuál era el día de la semana ℓ días atrás para este mismo producto.
  * Permite al modelo entender patrones semanales: por ejemplo, si hace 7 días (ℓ=7) fue domingo, ese valor de `día_de_la_semana_t-7 = 6` indica que el comportamiento de ventas de ese día de la semana se puede replicar o relacionar con la predicción actual.
  * Refuerza la información estacional: no solo “qué se vendió”, sino “en qué día de la semana se vendió”.

---

### 3. Lag Features de `mes`

Para cada entero $\ell = 1,2,\dots,14$:

* **Nombre de la columna**:

  ```
  * mes_t-ℓ
  ```

  (por ejemplo: `* mes_t-1`, `* mes_t-2`, …, `* mes_t-14`).

* **Fórmula**:

  ```python
  df[f"* mes_t-{ℓ}"] = df.groupby("product_id")["mes"].shift(ℓ)
  ```

  Donde `mes` ya se había definido como `df['dt'].dt.month` (1 = enero, …, 12 = diciembre).

* **Propósito**:

  * Indicar en qué mes estaba la fecha del registro ℓ días antes.
  * Permite capturar cambios de comportamiento ligados a meses contiguos o cruzar “día de la semana” con “mes” (p. ej., si hace 8 días fue 31 de marzo, y hoy es 8 de abril, el mes cambió; ese “cambio de mes” podría reflejar impacto en ventas).
  * Ayuda a modelar picos o valles mensuales de demanda (temporadas altas/bajas).

---

### 4. Lag Features de `es_fin_de_semana`

Para cada entero $\ell = 1,2,\dots,14$:

* **Nombre de la columna**:

  ```
  * es_fin_de_semana_t-ℓ
  ```

  (por ejemplo: `* es_fin_de_semana_t-1`, `* es_fin_de_semana_t-2`, …, `* es_fin_de_semana_t-14`).

* **Fórmula**:

  ```python
  df[f"* es_fin_de_semana_t-{ℓ}"] = df.groupby("product_id")["es_fin_de_semana"].shift(ℓ)
  ```

  Donde `es_fin_de_semana` es 1 si `día_de_la_semana` ∈ {5,6} (sábado/domingo), o 0 en otro caso.

* **Propósito**:

  * Señalar si, hace ℓ días, correspondía a fin de semana (valor 1) o no (valor 0).
  * Refuerza la influencia del “tipo de día” (laboral vs. fin de semana) en las ventas pasadas para que el modelo entienda, por ejemplo, que los sábados y domingos suelen tener comportamientos de demanda distintos.
  * Junto con los lags de `sale_amount`, permite discriminar si la baja/alta venta de hace ℓ días se debió a ser un fin de semana.

---

### 5. Lag Features de `ratio_sale_stock`

Para cada entero $\ell = 1,2,\dots,14$:

* **Nombre de la columna**:

  ```
  * ratio_sale_stock_t-ℓ
  ```

  (por ejemplo: `* ratio_sale_stock_t-1`, `* ratio_sale_stock_t-2`, …, `* ratio_sale_stock_t-14`).

* **Fórmula**:

  ```python
  df[f"* ratio_sale_stock_t-{ℓ}"] = df.groupby("product_id")["ratio_sale_stock"].shift(ℓ)
  ```

  Donde se definió previamente:

  ```python
  df['ratio_sale_stock'] = df['sale_amount'] / df['stock_hour6_22_cnt'].replace({0: np.nan})
  df['ratio_sale_stock'] = df['ratio_sale_stock'].replace([np.inf, -np.inf, np.nan], 0.0)
  ```

* **Propósito**:

  * Indicar la proporción entre venta y stock de hace ℓ días.
  * Ayuda a medir la velocidad de agotamiento de inventario: si hace ℓ días `ratio_sale_stock` era alto, significa que ese día el producto se vendió rápido respecto a su stock disponible.
  * Permite al modelo identificar “días de rotura de stock” o “días de baja rotación”, lo cual complementa el pronóstico de demanda con información de disponibilidad.

---

### 6. Lag Features de categorías One-Hot (`first_category_id` y `second_category_id`)

Para las categorías de primer nivel:

* Etiquetas posibles: `cat1_C1`, `cat1_C2`, …, `cat1_C5`, `cat1_OTRO` (donde C1–C5 son los 5 IDs más frecuentes de `first_category_id`).

  Para cada combinación de (Categoría, ℓ) con ℓ = 1…14:

  * **Nombre de la columna**:

    ```
    cat1_<valor>_t-ℓ
    ```

    por ejemplo: `* cat1_16_t-1`, `* cat1_16_t-2`, …, `* cat1_OTRO_t-14`.

  * **Fórmula**:

    ```python
    df[f"cat1_{valor}_t-{ℓ}"] = df.groupby("product_id")[f"cat1_{valor}"].shift(ℓ)
    ```

    donde `cat1_<valor>` era 1 si `first_category_id` == `<valor>` (o “OTRO” si no está en top 5), sino 0.

  * **Propósito**:

    * Reflejar si hace ℓ días el producto pertenecía a esa categoría de nivel 1 (valor 1) o no (valor 0).
    * Ayuda al modelo a reconocer patrones de comportamiento según la categoría histórica del producto (aunque esa categoría normalmente no cambia en el tiempo, su repetición en los lags refuerza que el producto se mantuvo inalterado en su agrupación).
    * Utilizar esta información en conjunto con lags de ventas y estacionalidad para segmentar pronósticos por categoría.

Para las categorías de segundo nivel:

* Etiquetas posibles: `cat2_C1`, `cat2_C2`, …, `cat2_C5`, `cat2_OTRO` (5 IDs más frecuentes de `second_category_id` + “OTRO”).

  Y de forma análoga, para cada (Categoría, ℓ):

  * **Nombre de la columna**:

    ```
    cat2_<valor>_t-ℓ
    ```

    por ejemplo: `cat2_12_t-1`, `cat2_34_t-5`, `* cat2_OTRO_t-14`.

  * **Fórmula**:

    ```python
    df[f"cat2_{valor}_t-{ℓ}"] = df.groupby("product_id")[f"cat2_{valor}"].shift(ℓ)
    ```

    donde `cat2_<valor>` es la dummy de segundo nivel obtenida por `pd.get_dummies`.

  * **Propósito**:

    * Proporciona contexto adicional de subcategoría histórica.
    * Permite al modelo diferenciar, por ejemplo, que hace 3 días el artículo estaba en `cat2 = 34`, lo que podría influir en patrones de venta distintos a otros subgrupos.
    * Junto con los lags de nivel 1, enriquece la segmentación temporal por jerarquía de categorías.

---

### 7. Variable Objetivo: `target_sale_next`

* **Nombre de la columna**:

  ```
  target_sale_next
  ```

* **Fórmula**:

  ```python
  df["target_sale_next"] = df.groupby("product_id")["sale_amount"].shift(-1)
  ```

  Es decir, para cada registro actual de un producto, esta columna toma el valor de `sale_amount` correspondiente al **día siguiente** (ℓ = –1) de ese mismo `product_id`.

* **Propósito**:

  * Es la variable que el modelo intentará predecir. Representa la **venta real del siguiente día** después de la ventana de 14 días de lag features.
  * Permite formular el problema como supervisado: dados los 14 lags de ventas, estacionalidad y demás features, predecir `sale_amount` del día siguiente.
  * En el entrenamiento, el algoritmo utiliza todas las demás columnas (lags y características temporales) como entradas (X) y `target_sale_next` como etiqueta (y).

---

## 8. Notas finales sobre el uso de lag features

1. **Eliminación de filas con NaN**

   * Después de generar todos los lags (ℓ = 1…14) para cada variable, quedaron muchos valores `NaN` en las primeras 14 filas por cada grupo de producto (porque no hay “ℓ días atrás”).
   * Se aplicó:

     ```python
     lag_columns = [
         "* sale_t-1", "* sale_t-2", …, "* sale_t-14",
         "día_de_la_semana_t-1", …, "día_de_la_semana_t-14",
         "* mes_t-1", …, "* mes_t-14",
         "* es_fin_de_semana_t-1", …, "* es_fin_de_semana_t-14",
         "* ratio_sale_stock_t-1", …, "* ratio_sale_stock_t-14",
         "cat1_…_t-1", …, "cat1_…_t-14",
         "cat2_…_t-1", …, "cat2_…_t-14"
     ]
     df_secuencias = df.dropna(subset=lag_columns + ["target_sale_next"])
     ```
   * Esto asegura que **todas** las filas del DataFrame final (`df_secuencias`) contengan valores válidos en cada uno de los 14 lags y en el target del día siguiente.

2. **Orden de columnas en `df_secuencias`**

   * First:

     1. `product_id`
     2. `fecha_final_ventana`
     3. `target_sale_next`
   * Luego:

     * `* sale_t-1`, `* sale_t-2`, …, `* sale_t-14`
     * `día_de_la_semana_t-1`, …, `día_de_la_semana_t-14`
     * `* mes_t-1`, …, `* mes_t-14`
     * `* es_fin_de_semana_t-1`, …, `* es_fin_de_semana_t-14`
     * `* ratio_sale_stock_t-1`, …, `* ratio_sale_stock_t-14`
     * Para cada valor de top 5 + “OTRO” en categorías de nivel 1: `cat1_<valor>_t-1`, …, `cat1_<valor>_t-14`
     * Para cada valor de top 5 + “OTRO” en categorías de nivel 2: `cat2_<valor>_t-1`, …, `cat2_<valor>_t-14`

3. **Importancia de cada grupo de features**

   * **Lags de ventas (`* sale_t-ℓ`)**: core de predicción, reflejan dinámica histórica de demanda.
   * **Lags de estacionalidad (`día_de_la_semana_t-ℓ`, `* mes_t-ℓ`, `* es_fin_de_semana_t-ℓ`)**: capturan patrones temporales repetitivos (por hora, día, mes, fines de semana).
   * **Lags de indicadores de stock (`* ratio_sale_stock_t-ℓ`)**: contextualizan la disponibilidad previa y posibles rupturas de stock.
   * **Lags de categorías one-hot**: ayudan a segmentar el comportamiento histórico según pertenencia a categoría y subcategoría.
   * **`target_sale_next`**: etiqueta real que comprobaremos al validar el modelo (predicción vs. valor observado).

---

> **Resumen**:
>
> * Existen **14 lags** para cada variable relevante (ventas, estacionalidad, rotación de stock, categorías).
> * Cada lag se obtuvo con `groupby("product_id").shift(ℓ)`.
> * El **target** `target_sale_next` se obtuvo con `shift(-1)` para predecir la venta del día siguiente.
> * En conjunto, estos features permiten al algoritmo “mirar hacia atrás” hasta 14 días y utilizar patrones de demanda, estacionalidad y stock para anticipar la venta en el futuro inmediato.
